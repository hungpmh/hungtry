## Scope

### Global scope vs local scope?

```python
var = 2

def multByVar(x):
    return x * var

print(multByVar(7))
print(var) 
```

```python
def multByVar(x):
    num = 10
    return x * num

print(multByVar(7))
print(num)
```

### `global` keyword
- A variable defined inside of a function is local by default
- use `global` keyword to read and write a global variable inside a function

```python
n = 10
def multByVar(x):
    global n
    n = 2
    return x * n

print(multByVar(7))
print(n)
```

```python
var = 2

def multByVar(x):
    return x * var, x**var, x*x
z = multByVar(7)
z1, _, z2 = z
print(z1, _)
print(var, _) 
```

### Scope Quiz

**What are results of these following lines of code?**

```python
a = 1

def fun():
    a = 2
    print(a)

fun()
print(a)
```
