```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('drug200.csv')
df.head()
```

```python
df.plot(x ='Age', y='Na_to_K', kind = 'scatter', figsize=(20,10))
```

```python
df_group = df.groupby('Sex').count()
df_group.head()
```

```python
df_group.plot.pie(y='Age',figsize=(5, 5),autopct='%1.1f%%', startangle=90)
```

```python
df.groupby('Cholesterol').count().plot.bar(y='Na_to_K', figsize=(20,10))
```

```python
df.describe()
```

```python
df.describe().boxplot(rot=45, fontsize=12, figsize=(8,10))
```

```python
df.groupby('Cholesterol').boxplot(rot=45, fontsize=12, figsize=(8,10))
```

```python
df_stock = pd.read_csv('TSLA.csv', parse_dates=['Date'])
df_stock.head()
```

```python
df_stock.plot.line(x='Date', y=['Open', 'High', 'Low', 'Close'], figsize = (20,10))
```

```python

df_stock['MA15'] = df_stock['Close'].rolling(15).mean()
df_stock['MA20'] = df_stock['Close'].rolling(20).mean()
df_stock.plot.line(x='Date', y=['Close', 'MA15', 'MA20'], figsize = (20,10))
```
