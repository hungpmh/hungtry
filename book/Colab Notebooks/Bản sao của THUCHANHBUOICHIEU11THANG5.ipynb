```python
def median(v): # Hàm xác định giá trị trung vị
 
   n = len(v)
 
   # sắp xếp tập mẫu
 
   sorted_v = sorted(v)
 
   midpoint = n // 2
 
   if n % 2 == 1:
 
     # nếu số phần tử của tập mẫu là lẻ thì Median là phần tử ở giữa sau khi
 
     # tập mẫu được sắp xếp
 
     return sorted_v[midpoint]
 
   else:
 
     # nếu số phần tử của tập mẫu là chẵn thì Median là Median của hai phần tử
 
     # ở giữa sau khi tập mẫu được sắp xếp
 
     lo = midpoint - 1
 
     hi = midpoint
 
     return (sorted_v[lo] + sorted_v[hi]) / 2

```

```python
print(median(So_giay))
```

```python
def quantile(x, p): # Hàm phân vị mức p: ứng với vị trí xác suất bằng p, hiển thị giá trị x0 sao cho xác suất nhận giá trị nhỏ hơn x0 sẽ bằng p
 
    p_index = int(p * len(x))
 
    return sorted(x)[p_index]
```

```python
from collections import Counter
 
S1 = [1, 1, 3, 3, 3, 4]
 
S2 = [1, 2, 3]
 
S3 = [1, 2, 2,2,3,4,4,4,5,4, 1]
 
def mode(x):
 
   counts = Counter(x)
 
   max_count = max(counts.values())
 
   return [x_i for x_i, count in counts.items()
 
             if count == max_count]
 
print(mode(S1)) # [3]
 
print(mode(S2)) # [1,2,3]
 
print(mode(S3))#[1,2]
```

```python
# Tính tổng bình phương
 
def sum_of_squares(s):
 
   return sum(s_i * s_i
 
          for s_i, s_i in zip(s, s))
 
# Định nghĩa hàm tính mean
 
def mean(s):
 
   return sum(s)/len(s)
 
# tính độ lệch
 
def deviation(s):
 
   s_Mean = mean(s)
 
   return [s_i - s_Mean for s_i in s]
 
# tính phương sai
 
def variance(s):
 
   n = len(s)
 
   d = deviation(s)
 
   return sum_of_squares(d) / (n - 1)
 
# tính độ lệch chuẩn
 
def standard_deviation(s):
 
  return np.sqrt(variance(s))
```

```python
def dot(x,y):
 
  return sum(x_i * y_i
 
       for x_i, y_i in zip(x, y))
 
# hiệp phương sai
 
def covariance(x, y):
 
  n = len(x)
 
  return dot(deviation(x), deviation(y)) / (n - 1)
 
# tính hệ số tương quan
 
def correlation(x, y):
 
   stdev_x = standard_deviation(x)
 
   stdev_y = standard_deviation(y)
 
   if stdev_x > 0 and stdev_y > 0:
 
      return covariance(x, y) / (stdev_x * stdev_y)
 
   else:
 
      return 0
```
