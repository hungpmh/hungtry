```python
import pandas as pd
Thinh= pd.read_csv('2008_swing_states.csv')# Đọc dữ liệu lưu trong đối tượng có tên là Thinh
Thinh.head(5)
```

```python
# Histogram là xấp xỉ hàm mật độ thực nghiệm
import matplotlib.pyplot as plt #Gọi thư viện matplotlib
plt.figure(figsize=(10,6))
_ = plt.hist(Thinh['dem_share']) # Vẽ biểu đồ dạng histogram cho cột dem_share: tỉ lệ vote cho đảng dân chủ (Obama)
_ = plt.xlabel('Tỉ lệ bầu cho Obama')#Tiêu đề trên trục hoành
_ = plt.ylabel('Số lượng các hạt')
plt.show()
```

```python
_ = plt.hist(Thinh['dem_share'], bins=40)
```

```python
import seaborn as sns # gọi thư viện seaborn
sns.set() #Thiết lập môi trường seaborn
_ = plt.hist(Thinh['dem_share']) # Vẽ biểu đồ dạng histogram cho dem_share
_ = plt.xlabel('percent of vote for Obama')
_ = plt.ylabel('number of counties')
plt.show()
```

Hình dáng của histogram cho ta xấp xỉ phân bố xác suất của tỉ lệ phiếu bầu cho Obama

```python
plt.figure(figsize=(14,6))
_ = sns.swarmplot(x='state', y='dem_share', data=Thinh) # Vẽ cho từng bang, mỗi giá trị của dem_share trong mỗi hạt của từng bang
# được xếp thành từng điểm không trùng nhau. Nếu 2 giá trị bằng nhau thì được xếp cạnh nhau và song song với nhau: cùng hàng ngang
_ = plt.xlabel('state')
_ = plt.ylabel('percent of vote for Obama')
plt.show()
```

```python
data= pd.read_csv('/content/2008_all_states.csv')# Đọc dữ liệu lưu trong đối tượng có tên là Thinh
data.head(5)
```

```python
plt.figure(figsize=(14,6))
_ = sns.swarmplot(x='east_west', y='dem_share', data=data) # Vẽ cho từng bang, mỗi giá trị của dem_share trong mỗi hạt của từng bang
# được xếp thành từng điểm không trùng nhau. Nếu 2 giá trị bằng nhau thì được xếp cạnh nhau và song song với nhau: cùng hàng ngang
_ = plt.xlabel('east_west')
_ = plt.ylabel('percent of vote for Obama')
plt.show()
```

```python
# Vẽ hàm phân bố thực nghiệm tích lũy
import numpy as np # Gọi thư viện numpy bao gồm các tính toán số học và đại số
x = np.sort(Thinh['dem_share'])
y = np.arange(1, len(x)+1) / len(x)
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('Tỉ lệ phần trăm phiếu bầu cho Obama')
_ = plt.ylabel('Xác suất tương ứng-Phân bố tích lũy')
plt.margins(0.02) # Keeps data off plot edges
plt.show()
```
