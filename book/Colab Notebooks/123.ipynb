```python
import pandas as pd
from pandas_datareader import data as web
import plotly.graph_objects as go
```

```python
stock = 'TSLA'
 
df = web.DataReader(stock, data_source='yahoo', start='09-01-2020')
```

```python
df
```

```python
df['Close'].plot()
```

```python
trace1 = {
   'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low,
    'type': 'candlestick',
    'name': 'TSLA',
    'showlegend': True
}
layout = go.Layout({
    'title': {
        'text': 'Đồ thi TSLA',
        'font': {
            'size': 15
        }
    }
})
```

```python
fig = go.Figure(data=trace1, layout=layout)
#fig.write_html("Microsoft(MSFT) Moving Averages.html")
fig.show()
```

```python
# Calculate and define moving average of 30 periods
avg_30 = df.Close.rolling(window=30, min_periods=1).mean()

# Calculate and define moving average of 50 periods
avg_50 = df.Close.rolling(window=50, min_periods=1).mean()

trace2 = {
    'x': df.index,
    'y': avg_30,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'blue'
            },
    'name': 'Moving Average of 30 periods'
}

trace3 = {
    'x': df.index,
    'y': avg_50,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'red'
    },
    'name': 'Moving Average of 50 periods'
}
```

```python
data=[trace1,trace2]
fig = go.Figure(data=data, layout=layout)
#fig.write_html("Microsoft(MSFT) Moving Averages.html")
fig.show()
```

```python
data = [trace1,trace2,trace3]
# Config graph layout
layout = go.Layout({
    'title': {
        'text': 'TSLA Moving Averages',
        'font': {
            'size': 15
        }
    }
})
```

```python
fig = go.Figure(data=data, layout=layout)
#fig.write_html("Microsoft(MSFT) Moving Averages.html")
fig.show()
```
