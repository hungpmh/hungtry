```python
#Cleanning data
all_columns = ['mainroad','guestroom', 'basement', 'hotwaterheating', 'airconditioning',
        'prefarea']
def binarymapping(x):
    return x.map({'yes':1 ,'no':0})

dataset[all_columns] = dataset[all_columns].apply(binarymapping)
dataset.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
fstatus = OneHotEncoder(sparse=False, handle_unknown='error', drop='first')
fstatus_dataset = pd.DataFrame(fstatus.fit_transform(dataset[['furnishingstatus']]))
fstatus_dataset.columns = fstatus.get_feature_names(['status'])
fstatus_dataset.head()
```

```python
dataset = pd.concat([dataset, fstatus_dataset], axis = 1)
dataset.drop(['furnishingstatus'], axis = 1, inplace = True)
dataset.head()
```

```python
#define the input [x] and the output / prediction data
#define the y first 
y=dataset['price']
# now the input data 
X=dataset.drop(['price'], axis=1)
```

```python
y = dataset.iloc[:,0].values 
y = y.reshape(-1, 1)
```

```python
dataset.info
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict([[5000,2,0,0,4,0,0,1,0,0,0,0.0,0]])
print(y_pred)
```

```python
sns.pairplot(dataset)
plt.show()
```

```python
viz_train = plt
viz_train.scatter(X_train, y_train, color='red')
viz_train.plot(X_train, regressor.predict(X_train), color='blue')
viz_train.title('Area VS Price (Training set)')
viz_train.xlabel('Area')
viz_train.ylabel('Price')
viz_train.show()
```
